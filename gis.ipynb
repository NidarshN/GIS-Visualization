```python
import networkx as nx
import osmnx as ox
import matplotlib.pyplot as plt
import geopandas as gpd
%matplotlib inline
ox.__version__
```

```python
polygon = gpd.read_file(r"data/birmingham/POLYGON.shp")
polygon_graph = ox.graph_from_polygon(polygon["geometry"].iloc[0], network_type="drive")
```

```python
fig, ax = ox.plot_graph(polygon_graph, node_size=0, bgcolor='white', edge_color='#143d59', edge_linewidth=0.5, dpi=1080, show=False, close=False)
ax.set_title('Birmingham, Alabama, USA')
plt.savefig('./images/birmingham.jpg', dpi=1080)
```
